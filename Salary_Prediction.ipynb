#**Salary Prediction Model**
A simple Linear Regression Model


```python
#Step 1 : import libraries
import pandas as pd
import numpy as np
```

```python
#step 2 : Read the data
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
#Step 3: Define Y (OUTPUT) and X(INPUT)
df.columns
```

```python
y = df['Salary'] #y is output or outcome that we want to Predict
x = df[['Experience Years']]  #x is the variable that we give as an input
```

```python
df.shape
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y, random_state=2526)
```

```python
#Step 4: select model
import seaborn as sns
sns.regplot(x=)
```

Equation of line
y = c + mx
where c= intercept means value of y at x = 0 , and slope is how much y exceeds when x exceeds.

In machine learning, y = c + mx
where y = output = c = b0 and mx b1
Best fit line = the line which gives me least sum of error sq, e^2
error = actual - prediction

```python
#step 5 : Select model
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#step 6 : train model
model.fit(x_train, y_train)
```

```python
model.intercept_ #it will give the intercept
```

```python
model.coef_ #it will give the scope
```

Equation of Best fit line

Salary = 27690 + 9328 * exp

```python
#step 7 : predict
y_pred = model.predict(x_test)
```

```python
#step 8 Evaluate
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

So the accuracy of this model is 0.084 * 100, i.e 8.4% of error rate or 91.6% acccuracy rate
